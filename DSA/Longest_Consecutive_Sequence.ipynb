```python
nums=[1,99,101,98,2,5,3,100,1,1]

max_count=0
n=len(nums)

for i in range (0,n):
    count=1
    num=nums[i]

    while num+1 in nums:
        count+=1
        num=num+1
    max_count=max(count,max_count)
print(max_count)




#better with help of sorting

numb=[1,99,101,98,2,5,3,100,1,1]
nums=numb.sort()
n=len(nums)
last_small=float("-infy")
longest=0

for i in range(0,n):
    count=1
    num=nums[i]
    if num-1==last_small:
        count+=1
        last_small=num
    elif num-1!=last_small:
        count=1
        last_small=num

    longest=max(longest,count)

print(longest)

        

# optimal Sol^n

numb=[1,99,101,98,2,5,3,100,1,1]
myset=set()
longest=0
for i in range (0,n):
    myset.add(nums[i]))

for num in myset:
    if num-1 not in myset:
        count=1
        x=num
        while num+1 in myset:
            count+=1
            x+=1

        longest=max(longest,count)




    
```
