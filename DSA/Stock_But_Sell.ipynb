# 📈 Best Time to Buy and Sell Stock

#  Brute Force (While Loop)

## 🧠 Idea

- Try every pair `(buy, sell)`  
- Profit = `prices[j] - prices[i]`  
- Track maximum profit  

---

````python
def max_profit_brute_while(prices):
    n = len(prices)
    maxi = 0

    for i in range(n):
        j = i + 1
        while j < n:
            profit = prices[j] - prices[i]
            maxi = max(maxi, profit)
            j += 1

    return maxi
````

````python
prices = [7,2,1,5,6,4,8]
max_profit_brute_while(prices)
````

## ⏱️ Complexity

**Time: O(n²)**

* Checking all pairs

**Space: O(1)**

---

## 🧩 Pattern

* Nested Loop

#  Brute Force (For Loop)

## 🧠 Idea

* Same as above but cleaner
* Only check if selling price > buying price

---

````python
def max_profit_brute_for(prices):
    n = len(prices)
    maxi = 0

    for i in range(n):
        for j in range(i+1, n):
            if prices[j] > prices[i]:
                profit = prices[j] - prices[i]
                maxi = max(maxi, profit)

    return maxi
````

````python
prices = [7,2,1,5,6,4,8]
max_profit_brute_for(prices)
````

# 🔥 Optimal (Greedy)

## 🧠 Idea

* Keep track of **minimum price so far**
* At each step:

  ```
  profit = current price - min_price
  ```
* Update max profit

---

````python
def max_profit_optimal(prices):
    min_price = float("inf")
    max_profit = 0

    for price in prices:
        min_price = min(min_price, price)
        profit = price - min_price
        max_profit = max(max_profit, profit)

    return max_profit
````

````python
prices = [7,2,1,5,6,4,8]
max_profit_optimal(prices)
````

## ⏱️ Complexity

**Time: O(n)**

* Single pass

**Space: O(1)**

---

## 🧩 Pattern

* Greedy
* Tracking minimum

---

## 🎯 When to Use

* Buy-sell problems
* Max profit / difference problems
* Single transaction

---

## ⚠️ Notes

* Buy before sell (important constraint)
* Track min BEFORE computing profit
