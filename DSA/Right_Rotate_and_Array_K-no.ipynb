# 🔁 Rotate Array (Right Rotation by k)

## 🧠 Problem
Given an array `nums` and integer `k`, rotate the array to the right by `k` steps.

### Example
```

nums = [3, 9, 5, 6, 7, 2]
k = 3

Output:
[6, 7, 2, 3, 9, 5]

```

---

# 🧠 1. Idea / Intuition

## Brute Force
- Rotate array one step at a time
- Repeat `k` times

### Steps:
1. Remove last element
2. Insert at front
3. Repeat k times

`````python
def rotate(nums, k):
    n = len(nums)
    k = k % n

    for _ in range(k):
        last = nums.pop()
        nums.insert(0, last)
`````

`````python
nums = [3, 9, 5, 6, 7, 2]
k = 3
rotate(nums, k)
print(nums)
`````

## Complexity

### Brute Force

* Time: O(n * k)

  * Each rotation takes O(n)
* Space: O(1)

## 🔥 Optimal Approach (Reverse Trick)

### Steps:
1. Reverse last `k` elements
2. Reverse first `n-k` elements
3. Reverse entire array

---

`````python
def rotate(nums, k):
    n = len(nums)
    k = k % n

    def reverse(left, right):
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1

    reverse(n-k, n-1)   # last k
    reverse(0, n-k-1)   # first n-k
    reverse(0, n-1)     # whole array
`````

`````python
nums = [3, 9, 5, 6, 7, 2]
k = 3
n = 6
rotate(nums, k)
print(nums)
`````

# ⏱️ Complexity

* Time: O(n)

  * Each element visited ~3 times
* Space: O(1)

  * In-place swaps

---

## ⚡ Better (Slicing)

### Insight:
Last `k` elements move to front

```

nums = [3, 9, 5, 6, 7, 2]
k = 3

Last k → [6, 7, 2]
Remaining → [3, 9, 5]

Result → [6, 7, 2, 3, 9, 5]

````


`````python
def rotate(nums, k):
    n = len(nums)
    k = k % n

    nums[:] = nums[n-k:] + nums[:n-k]
`````

`````python
nums = [3, 9, 5, 6, 7, 2]
k = 3
rotate(nums, k)
print(nums)
`````

## Complexity

## Slicing

* Time: O(n)

  * Copying two parts
* Space: O(n)

# 🧩  Patterns Used

* Two Pointers (for reversing)
* Array Manipulation
* In-place algorithm

---

# 🎯  When to Use This

### Use Reverse Trick when:

* Problem says **"in-place"**
* Rotation or rearrangement required
* Space optimization is important

### Use Slicing when:

* Space is allowed
* Need quick and readable solution

# 🚨 Interview Tips

* Always do `k = k % n`

  * Handles large k
* Clarify:

  * Left vs Right rotation
* Mention brute → better → optimal
