# 🎯 Floor & Ceil in Sorted Array

---



## 🧠 Problem Statement

Given a **sorted array `nums`** and a `target`:

### 🔹 Floor
👉 Largest number **≤ target**

### 🔹 Ceil
👉 Smallest number **≥ target**

---

## 🧪 Input

```python
nums = [3,4,4,4,8,9,9,10,12,12,14,15]
target = 11
````

---

## ✅ Output

```python
floor = 10
ceil  = 12
```

---

## 📦 Visual

```
nums = [3,4,4,4,8,9,9,10,12,12,14,15]
                      ↑
                   target = 11
```

---

# 🧠 0. INTUITION

## 💡 Idea

Think in boundaries:

```
[ <= target | > target ]
      ↑ floor    ↑ ceil
```

---

## 🔥 Key Insight

👉 Floor = last valid on LEFT
👉 Ceil  = first valid on RIGHT

---



# 🧱 1. BRUTE FORCE (Linear Search)

## 💡 Idea

👉 Traverse entire array and update floor & ceil

👉 Pattern: **Linear Scan**

---

`````python
def floorCeil(nums, target):
    floor = -1
    ceil = -1

    for num in nums:
        if num <= target:
            floor = num   # keep updating (last valid)
        if num >= target and ceil == -1:
            ceil = num    # first valid

    return [floor, ceil]
`````

`````python
nums = [3,4,4,4,8,9,9,10,12,12,14,15]
floorCeil(nums, 11)
`````

## 🔍 Dry Run

### target = 11

```
num=3 → floor=3
num=4 → floor=4
...
num=10 → floor=10
num=12 → ceil=12 (first ≥ target)
```

👉 Answer = [10, 12]

---

## ⏱ Complexity

### Time: `O(n)`

👉 full traversal

### Space: `O(1)`

---

## 🚨 When to Use

* Small constraints
* No need for optimization

---



# 🚀 2. OPTIMAL (Binary Search)

## 💡 Idea

👉 Combine:

* Floor → move right when valid
* Ceil → move left when valid

👉 Pattern: **Binary Search (Boundary Tracking)**

---

`````python
def floorCeil(nums, target):
    low, high = 0, len(nums) - 1
    floor = -1
    ceil = -1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == target:
            return [nums[mid], nums[mid]]

        elif nums[mid] < target:
            floor = nums[mid]   # valid floor
            low = mid + 1

        else:
            ceil = nums[mid]    # valid ceil
            high = mid - 1

    return [floor, ceil]
`````

`````python
nums = [3,4,4,4,8,9,9,10,12,12,14,15]
floorCeil(nums, 11)
`````

## 🔍 Dry Run

### target = 11

```
low=0, high=11
mid=5 → 9 <11 → floor=9 → right

low=6, high=11
mid=8 → 12 >11 → ceil=12 → left

low=6, high=7
mid=6 → 9 <11 → floor=9 → right

low=7, high=7
mid=7 → 10 <11 → floor=10 → right

low=8, high=7 → stop
```

👉 Answer = [10, 12]

---

## ⏱ Complexity

### Time: `O(log n)`

👉 halving search space

### Space: `O(1)`

---

# ❓ WHY floor = -1 / ceil = -1 ?

## 💡 Very Important Edge Case Logic

We initialize:

```python
floor = -1
ceil = -1
```

👉 Means:

```
"Not found yet"
```

---

## 🧠 Case 1: No Floor Exists

### Example:

```python
nums = [3,4,5]
target = 2
```

```
No number ≤ 2
→ floor = -1
→ ceil = 3
```

---

## 🧠 Case 2: No Ceil Exists

### Example:

```python
nums = [3,4,5]
target = 10
```

```
No number ≥ 10
→ floor = 5
→ ceil = -1
```

---

## 🔥 Why NOT 0?

👉 Because 0 can be a valid array value

👉 So we use:

```
-1 = "does not exist"
```

# 🧠 Pattern Summary

| Concept | Direction  |
| ------- | ---------- |
| Floor   | move RIGHT |
| Ceil    | move LEFT  |

# 🔥 Interview Questions They Ask

## 1️⃣ Why binary search works here?

👉 Because array is sorted → we can eliminate half

---

## 2️⃣ Difference between LB and Ceil?

```
Lower Bound = index
Ceil = value
```

---

## 3️⃣ Can we solve using LB?

👉 YES:

```python
ceil = nums[lowerBound(nums, target)]
floor = nums[lowerBound(nums, target) - 1]
```

---

## 4️⃣ What if LB = 0?

👉 Then:

```
No floor exists
```
