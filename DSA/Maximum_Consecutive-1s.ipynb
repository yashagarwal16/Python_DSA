# 🔢 Maximum Consecutive 1s

---

## 🧠 Idea

- Count continuous `1`s  
- Reset count when `0` appears  
- Track maximum streak  

---

````python
def max_consecutive_ones(nums):
    count = 0
    max_count = 0

    for num in nums:
        if num == 1:
            count += 1
        else:
            max_count = max(max_count, count)
            count = 0

    max_count = max(max_count, count)

    return max_count
````

````python
nums = [1,1,0,1,1,1]

max_consecutive_ones(nums)
````

## ⏱️ Complexity

**Time: O(n)**

* Single traversal

**Space: O(1)**

* Only variables used

---

## 🧩 Pattern

* Sliding Window (fixed condition)
* Counting

---

## 🎯 When to Use

* Longest streak / consecutive elements
* Binary arrays (0/1 problems)
* Subarray with condition
