# 🧮 Set Matrix Zeroes

---

# ❌ Brute Force

## 🧠 Idea

- If any cell is `0`  
- Mark its entire **row & column**  
- Use `inf` as temporary marker  
- Later convert `inf → 0`  

---


````python
def mark_infinity(matrix, row, col):
    n = len(matrix)
    m = len(matrix[0])

    # mark row
    for j in range(m):
        if matrix[row][j] != 0:
            matrix[row][j] = float('inf')

    # mark column
    for i in range(n):
        if matrix[i][col] != 0:
            matrix[i][col] = float('inf')
````

````python
def set_zero_brute(matrix):
    n = len(matrix)
    m = len(matrix[0])

    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 0:
                mark_infinity(matrix, i, j) ## Calling Function

    for i in range(n):
        for j in range(m):
            if matrix[i][j] == float('inf'):
                matrix[i][j] = 0

    return matrix
````

````python
matrix = [
 [7,9,2,3],
 [20,8,0,10],
 [29,0,-10,5],
 [4,14,6,7]
]

set_zero_brute(matrix)
````

# ⚡ Better (Using Extra Arrays)

## 🧠 Idea

* Track which rows & cols need zero
* Use arrays

---

````python
def set_zero_better(matrix):
    n = len(matrix)
    m = len(matrix[0])

    rowtrack = [0]*n
    coltrack = [0]*m

    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 0:
                rowtrack[i] = 1
                coltrack[j] = 1

    for i in range(n):
        for j in range(m):
            if rowtrack[i] == 1 or coltrack[j] == 1:
                matrix[i][j] = 0

    return matrix
````

````python
matrix = [
 [7,9,2,3],
 [20,8,0,10],
 [29,0,-10,5],
 [4,14,6,7]
]

set_zero_better(matrix)
````

# 🔥 Optimal (Constant Space)

## 🧠 Idea

* Use first row & column as markers
* Avoid extra space

---

````python
def set_zero_optimal(matrix):
    n = len(matrix)
    m = len(matrix[0])

    col0 = 1

    # Step 1: mark
    for i in range(n):
        if matrix[i][0] == 0:
            col0 = 0
        for j in range(1, m):
            if matrix[i][j] == 0:
                matrix[i][0] = 0
                matrix[0][j] = 0

    # Step 2: fill
    for i in range(1, n):
        for j in range(1, m):
            if matrix[i][0] == 0 or matrix[0][j] == 0:
                matrix[i][j] = 0

    # Step 3: first row
    if matrix[0][0] == 0:
        for j in range(m):
            matrix[0][j] = 0

    # Step 4: first col
    if col0 == 0:
        for i in range(n):
            matrix[i][0] = 0

    return matrix
````

````python
matrix = [
 [7,9,2,3],
 [20,8,0,10],
 [29,0,-10,5],
 [4,14,6,7]
]

set_zero_optimal(matrix)
````
