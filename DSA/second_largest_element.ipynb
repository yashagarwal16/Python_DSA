# 🔍 Second Largest Element (Without Sorting)

## 📌 Problem
Find the **second largest element** in an array without sorting.

```python
nums = [55, 32, 97, -55, 45, 32, 88, 21]

````python
# 🟡 Approach 1: Two Pass (Better but not optimal)
# 🧠 Idea
# First pass → find largest
# Second pass → find second largest

def second_largest(nums):
    largest = float("-inf")

    # First pass: find largest
    for num in nums:
        largest = max(largest, num)

    second_largest = float("-inf")

    # Second pass: find second largest
    for num in nums:
        if num != largest and num > second_largest:
            second_largest = num

    return second_largest
````

````python
# ⚡ Complexity
# Time Complexity → O(n + n) = O(2n) ≈ O(n)
# Space Complexity → O(1)
# ❌ Drawback
# Two loops → slightly less efficient
````

````python
# Approach 2: Optimal (Single Pass)
# 🧠 Idea
# Maintain:
# largest
# second_largest
# Update both in one traversal



def second_largest(nums):
    largest = float("-inf")
    second_largest = float("-inf")

    for num in nums:
        if num > largest:
            second_largest = largest
            largest = num
        
        elif num > second_largest and num != largest:
            second_largest = num

    return second_largest
````


## 🔄 Dry Run

```python
nums = [55, 32, 97, -55, 45, 32, 88, 21]
```

| Step | num | largest | second_largest |

| 2    | 32  | 55      | 32             |

| 7    | 88  | 97      | 88             |

👉 Final Answer = **88**

# ⚠️ Edge Cases

```python
[5] → No second largest ❌
[5, 5, 5] → No second largest ❌
[-1, -2, -3] → Works ✅
```
