# 🧮 Matrix (2D List) — Complete Notes

---

## 🧠 Idea

- A **matrix = list of lists**
- Each inner list = **row**
- Elements accessed using:
```

nums[i][j]

````

---

## 📌 Representation

```python
nums = [
  [5, 20, 3],
  [7, -10, 9],
  [1, -52, 6]
]
````

---

## 📏 Dimensions

```python
rows = len(nums)
cols = len(nums[0])
```

---



# 🔁 Traversal

## 🧠 Idea

* Outer loop → rows
* Inner loop → columns

---


`````python
nums = [
  [5, 20, 3],
  [7, -10, 9],
  [1, -52, 6]
]
rows = len(nums)
cols = len(nums[0])
`````

`````python
for i in range(len(nums)):
    for j in range(len(nums[0])):
        print(nums[i][j], end=" ")
    print()
`````

## 🔍 Dry Run

```text
5 20 3  
7 -10 9  
1 -52 6
```

---

## ⏱️ Complexity

**Time: O(rows × cols)**
**Space: O(1)**

---

## 🧩 Pattern

* Nested Loop

---



# 🔺 Upper Triangle

## 🧠 Idea

* Print when:

  ```
  j >= i
  ```

---

`````python
for i in range(len(nums)):
    for j in range(len(nums[0])):
        if j >= i:
            print(nums[i][j], end=" ")
        else:
            print("*", end=" ")
    print()
`````

## 🔍 Output

```text
5 20 3  
* -10 9  
* * 6
```

---




# 🔻 Lower Triangle

## 🧠 Idea

* Print when:

  ```
  j <= i
  ```

---



`````python
for i in range(len(nums)):
    for j in range(len(nums[0])):
        if j <= i:
            print(nums[i][j], end=" ")
        else:
            print("*", end=" ")
    print()
`````

# 🔷 Diagonal

## 🧠 Idea

* Main diagonal:

  ```
  i == j
  ```

---

`````python
for i in range(len(nums)):
    for j in range(len(nums[0])):
        if i == j:
            print(nums[i][j], end=" ")
        else:
            print("*", end=" ")
    print()
`````

# 🔷 Alternate Diagonal (Anti-Diagonal)

---

## 🧠 Idea

- Opposite of main diagonal  
- Condition:
```

i + j == n - 1

````

👉 Works for square matrix (n x n)

---

`````python
nums = [
  [5, 20, 3],
  [7, -10, 9],
  [1, -52, 6]
]

n = len(nums)

for i in range(n):
  for j in range(n):
      if i + j == n - 1:
          print(nums[i][j], end=" ")
`````

# 🔄 Transpose of Matrix

## 🧠 Idea

* Convert rows → columns
* Swap indices:

  ```
  result[j][i] = nums[i][j]
  ```

---

`````python
def transpose(nums):
    rows = len(nums)
    cols = len(nums[0])

    result = [[0]*rows for _ in range(cols)]

    for i in range(rows):
        for j in range(cols):
            result[j][i] = nums[i][j]

    return result
`````

`````python
nums = [
 [5,10,8],
 [7,6,3],
 [2,1,9]
]

transpose(nums)
`````

## ⏱️ Complexity

## 🧩 Pattern

* Matrix Manipulation

## 🎯 When to Use
