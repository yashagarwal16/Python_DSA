# 🔄 Rotate Matrix (90° Clockwise)


## 🧠 Idea

👉 Only works for **square matrix (n × n)**  

Goal:
- Rotate matrix by **90° clockwise**

---

## 🔁 Movement (Index Mapping)

```text
(i,j) → (j, n-1-i)

(0,0)=1 → (0,2)
(0,1)=2 → (1,2)
(0,2)=3 → (2,2)

(1,0)=4 → (0,1)
(1,1)=5 → (1,1)
(1,2)=6 → (2,1)

(2,0)=7 → (0,0)
(2,1)=8 → (1,0)
(2,2)=9 → (2,0)
````

# ❌ Brute Force

## 🧠 Idea

* Create new matrix
* Place elements using:

  ```
  result[j][n-1-i] = matrix[i][j]
  ```

---

`````python
def rotate_brute(matrix):
    n = len(matrix)
    result = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            result[j][n-1-i] = matrix[i][j]

    return result
`````

`````python
matrix = [
 [1,2,3],
 [4,5,6],
 [7,8,9]
]

rotate_brute(matrix)
`````

# 🔥 Optimal (Transpose + Reverse)

## 🧠 Idea

👉 Instead of new matrix:

1. Transpose
2. Reverse each row

---

`````python
def rotate_optimal(matrix):
    n = len(matrix)

    # Step 1: Transpose
    for i in range(n):
        for j in range(i+1, n):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

    # Step 2: Reverse rows
    for i in range(n):
        matrix[i].reverse()

    return matrix
`````

`````python
matrix = [
 [1,2,3],
 [4,5,6],
 [7,8,9]
]

rotate_optimal(matrix)
`````

---

## ⏱️ Complexity

**Time: O(n²)**

* Two passes

**Space: O(1)**

* In-place

---

## 🧩 Pattern

* Matrix transformation
* In-place swapping

---

## 🎯 When to Use

* Rotate image / matrix
* In-place constraint
* Square matrix problems

---

## ⚠️ Notes

* Works only for **n × n matrix**
* Transpose carefully (`j = i+1`)
* Reverse step is required
