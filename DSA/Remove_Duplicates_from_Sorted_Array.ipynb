# 🔁 Remove Duplicates from Sorted Array

## 📌 Problem
Given a **sorted array**, remove duplicates **in-place**  
Return the number of **unique elements**

---

## 🟡 Approach 1: Brute Force Method

### 🧠 Idea
- Store elements in a dictionary
- Copy unique elements back to array

---




````python
def remove_duplicates(nums):
    freq_map = {}

    # Store unique elements
    for num in nums:
        freq_map[num] = 1

    j = 0

    # Put back unique elements
    for key in freq_map:
        nums[j] = key
        j += 1

    return j
````

````python
nums=[1,1,1,3,3,3,4,5,6,7,7,8,8,8,9,9,9,9,9,9]
remove_duplicates
````

## ⚡ Complexity

* Time Complexity → **O(n + n) = O(n)**
* Space Complexity → **O(n)** ❌

---

## ❌ Drawback

* Extra space used
* Not in-place

---



# 🟢 Approach 2: Optimal (Two Pointers)

## 🧠 Idea

* Use two pointers:

  * `i` → last unique element
  * `j` → traversal pointer

👉 Since array is sorted, duplicates are adjacent

---



````python

def remove_duplicates(nums):
    n = len(nums)

    if n == 0:
        return 0

    i = 0

    for j in range(1, n):
        if nums[j] != nums[i]:
            i += 1
            nums[i] = nums[j]

    return i + 1

````


## 🔄 Dry Run

```python
nums = [1, 1, 2, 2, 3, 4]
```

| Step | i | j | nums             |
| ---- | - | - | ---------------- |
| Init | 0 | - | [1,1,2,2,3,4]    |
| j=1  | 0 | 1 | same (duplicate) |
| j=2  | 1 | 2 | [1,2,...]        |
| j=3  | 1 | 3 | skip             |
| j=4  | 2 | 4 | [1,2,3,...]      |
| j=5  | 3 | 5 | [1,2,3,4,...]    |

👉 Unique count = **4**

---

## ⚡ Complexity

* Time Complexity → **O(n)** ✅
* Space Complexity → **O(1)** ✅

---

## 💡 Key Insight

> Since array is sorted → duplicates are adjacent
> → So we don’t need extra space

---
