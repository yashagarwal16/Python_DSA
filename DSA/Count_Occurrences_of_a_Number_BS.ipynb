# 🎯 Count Occurrences of a Number

---



## 🧠 Problem Statement

Given a **sorted array `nums`** and a `target`:

👉 Return **number of times target appears**

---

## 🧪 Input

```python
nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
````

---

## ✅ Output

```python
5
```

---

# 🧠 0. INTUITION

## 💡 Idea

👉 Instead of counting one-by-one:

```
Count = last_index - first_index + 1
```

OR

```
Count = UpperBound - LowerBound
```

---



# 🧱 1. BRUTE FORCE

## 💡 Idea

👉 Count directly

👉 Pattern: **Linear Scan**

---

`````python
def countOccurrences(nums, target):
    count = 0

    for num in nums:
        if num == target:
            count += 1

    return count
`````

`````python
nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
countOccurrences(nums, target)
`````

## 🔍 Dry Run

```
target = 3
→ found at index 2,3,4,5,6
→ count = 5
```

---

## ⏱ Complexity

### Time: `O(n)`

👉 scan entire array

### Space: `O(1)`

---



# 🚀 2. OPTIMAL (Binary Search)

## 💡 Idea

👉 Use:

```
count = upperBound - lowerBound
```

---

# ❓ MAIN DOUBT: WHY UB = n ?

## 💡 THIS IS VERY IMPORTANT 🔥

### 👉 We initialize:

```python
ub = n
```

---

## 🧠 Meaning

```
"Assume answer is outside array (right side)"
```

👉 This handles edge cases cleanly

---

## 🔥 Think Like This

Upper Bound =
👉 first index where `nums[i] > target`

---

## ❗ Case 1: Target exists

```python
nums = [1,2,3,3,3,5]
target = 3
```

```
UB = index of 5 → 5
```

✔ works fine

---

## ❗ Case 2: Target is largest element

```python
nums = [1,2,3,3,3]
target = 3
```

👉 There is NO element > 3

So:

```
UB should be = n (out of array)
```

👉 That’s why we initialize:

```python
ub = n
```

---

## ❗ Case 3: Target not present

```python
nums = [1,2,4,5]
target = 3
```

```
LB = 2
UB = 2
count = 0
```

✔ still works

---

`````python
def lowerBound(nums, target):
    low, high = 0, len(nums) - 1
    lb = len(nums)

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] >= target:
            lb = mid
            high = mid - 1
        else:
            low = mid + 1

    return lb


def upperBound(nums, target):
    low, high = 0, len(nums) - 1
    ub = len(nums)   # ❗ important

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] > target:
            ub = mid
            high = mid - 1
        else:
            low = mid + 1

    return ub


def countOccurrencesOp(nums, target):
    lb = lowerBound(nums, target)
    ub = upperBound(nums, target)

    return ub - lb
`````

`````python
nums = [1,2,3,3,3,3,3,5,6,8,9,9,10]
target = 3
countOccurrencesOp(nums, target)
`````

## 🔍 Dry Run

### target = 3

```
LB = 2
UB = 7

count = 7 - 2 = 5 ✅
```

---

# ⏱ Complexity

### Time: `O(log n)`

👉 Two binary searches

### Space: `O(1)`

---

# 🧠 Pattern Summary

| Concept | Meaning        |
| ------- | -------------- |
| LB      | first ≥ target |
| UB      | first > target |
| Count   | UB - LB        |

---

# 🔥 Interview Insight (VERY IMPORTANT)

## Why NOT ub = -1 ?

👉 Because UB represents:

```
first index > target
```

👉 That can be:

```
inside array OR outside (n)
```

So:

```python
ub = n  ✔ correct default
```

---

# 💯 Final Intuition

```
LB → start of target block
UB → end of target block + 1

Count = size of block
```
