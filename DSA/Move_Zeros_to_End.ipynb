# 🚫 Move Zeros to End

---

## 🧠 Idea

- Keep all **non-zero elements in front**
- Push all **zeros to the end**

---


## 💻 Code (Brute - Extra Space)

`````python
def moveZeroes(nums):
    temp = []

    # store non-zero elements
    for num in nums:
        if num != 0:
            temp.append(num)

    # fill back into nums
    n2 = len(temp)

    for i in range(n2):
        nums[i] = temp[i]

    for i in range(n2, len(nums)):
        nums[i] = 0
`````

`````python
nums = [1, 2, 4, 0, 3, 0, 0, 3, 5, 1]
moveZeroes(nums)
print(nums)
`````

## 💻 Code (Optimal - Two Pointer)

`````python

def moveZeroes(nums):
    i = 0  # position for next non-zero

    for j in range(len(nums)):
        if nums[j] != 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
`````

`````python
nums = [1, 2, 4, 0, 3, 0, 0, 3, 5, 1]
moveZeroes(nums)
print(nums)
`````

## ⏱️ Complexity

### Brute

**Time: O(n)**

* One pass to collect + one to fill

**Space: O(n)**

* Extra temp array

### Optimal

**Time: O(n)**

* Single traversal

**Space: O(1)**

* In-place swaps

---

## 🧩 Pattern

* Two Pointer
* In-place array manipulation

---

## 🎯 When to Use

* When asked to **move elements while maintaining order**
* When problem says **in-place**
* When separating elements based on condition (0 vs non-zero)

---

## ⚠️ Notes

* Order of non-zero elements must be preserved
* Swap only when `nums[j] != 0`
