# 🔁 Right Rotate an Array by 1

## 📌 Problem
Rotate the array to the **right by 1 position**

```python
nums = [1, 4, 3, 2, 5]
````

### Output:

```python
[5, 1, 4, 3, 2]
```

---


# 🟡 Approach 1: Using Slicing (Python Shortcut)

`````python
nums = [1, 4, 3, 2, 5]

nums[:] = [nums[-1]] + nums[:-1]

print(nums)
`````

## 🧠 Explanation

* `nums[-1]` → last element
* `nums[:-1]` → all elements except last
* Combine → new rotated array

---

## ⚡ Complexity

* Time → **O(n)**
* Space → **O(n)** ❌ (new list created)

---

🟢 Approach 2: Without Slicing (Interview Preferred)

## 🧠 Idea

1. Store last element
2. Shift all elements to right
3. Place last element at index 0

---

`````python
def right_rotate(nums):
    n = len(nums)

    temp = nums[n - 1]   # store last element

    for i in range(n - 2, -1, -1):
        nums[i + 1] = nums[i]

    nums[0] = temp
`````

`````python
nums = [1, 4, 3, 2, 5]
right_rotate(nums)
`````

`````python
print(nums)
`````

## 🔄 Dry Run

```python
nums = [1, 4, 3, 2, 5]
```

### Step 1:

```text
temp = 5
```

### Step 2: Shift Right

```text
i=3 → [1,4,3,2,2]
i=2 → [1,4,3,3,2]
i=1 → [1,4,4,3,2]
i=0 → [1,1,4,3,2]
```

### Step 3:

```text
nums[0] = 5
```

### Final:

```text
[5, 1, 4, 3, 2]
```

---

## ⚡ Complexity

* Time → **O(n)** ✅
* Space → **O(1)** ✅

# ❌ Your Mistake (Fix)

```python
temp = nums[len(nums-1)] ❌
```

✅ Correct:

```python
temp = nums[len(nums) - 1]
```
