
# 🔗 Merge Two Sorted Arrays (Without Duplicates)

---

## 🧠 Idea

👉 Goal:
- Combine two arrays  
- Remove duplicates  
- Return sorted result  

---

# ✅ Case 1: Sorted Arrays

## 🧠 Idea

- Use **Two Pointer**
- Traverse both arrays together  
- Always pick **smaller element**
- Avoid duplicates using:
```

result[-1] != element

````

---



`````python
def union_sorted(nums, nums2):
  i = j = 0
  n1, n2 = len(nums), len(nums2)
  result = []

  while i < n1 and j < n2:
      if nums[i] <= nums2[j]:
          if not result or result[-1] != nums[i]:
              result.append(nums[i])
          i += 1
      else:
          if not result or result[-1] != nums2[j]:
              result.append(nums2[j])
          j += 1

  while i < n1:
      if not result or result[-1] != nums[i]:
          result.append(nums[i])
      i += 1

  while j < n2:
      if not result or result[-1] != nums2[j]:
          result.append(nums2[j])
      j += 1

  return result
`````
