# 🔍 Find the Largest Element in an Array

## 📌 Problem
Find the **maximum (largest) element** in a given array.

```python
nums = [55, 32, -97, 99, 3, 67]

if we put 
largest = 0
Problem:
Works only for positive numbers
Fails for:
nums = [-55, -32, -97, -99, -3, -67]

👉 Output will be 0 ❌ (incorrect)

````python
def find_largest(nums):
    largest = nums[0]

    for i in range(len(nums)):
        largest = max(largest, nums[i])

    return largest
````

````python
nums = [55, 32, -97, 99, 3, 67]

print(find_largest(nums))
````

🔄 Dry Run
nums = [55, 32, -97, 99, 3, 67]

Step	nums[i]	largest (before)	largest (after)
1	      55	  55	                   55
2	      32	  55	                   55
3	     -97	  55	                   55
4	      99	  55	                   99
5	      3	      99	                   99
6	      67	  99	                   99

👉 Final Answer = 99

````python
def find_largest(nums):
    largest = float("-inf")

    for i in range(len(nums)):
        largest = max(largest, nums[i])

    return largest
````

````python
nums = [55, 32, -97, 99, 3, 67]

print(find_largest(nums))
````

⚡ Complexity

Time Complexity → O(n)

Space Complexity → O(1)

❌ Don't initialize with 0 

✅ Use nums[0] or -∞

🔁 Traverse once → update max

🔥 Interview Tip

Always handle negative numbers when finding max/min
