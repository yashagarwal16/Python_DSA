# 🔍 Linear Search

---

## 🧠 Idea

- Traverse the array  
- Compare each element with `target`  
- Return index if found, else `-1`



````python
def linear_search(nums, target):
    n = len(nums)

    for i in range(n):
        if nums[i] == target:
            return i

    return -1
````

````python
nums=[1,2,3,2,3,4,5,3,2,10,9]
liner_search(nums,11)
````

## 🔍 Dry Run

```python
nums = [4, 2, 7, 1, 9]
target = 7
```

```
i=0 → nums[0]=4 ≠ 7  
i=1 → nums[1]=2 ≠ 7  
i=2 → nums[2]=7 == 7 → return 2
```

---

## ⏱️ Complexity

**Time: O(n)**

* Worst case: traverse entire array

**Space: O(1)**

* No extra space used

---

## 🧩 Pattern

* Simple Traversal
* Brute Force

---

## 🎯 When to Use

* Unsorted array
* Small input size
* No preprocessing allowed

---

## ⚠️ Notes

* Works on any array (sorted/unsorted)
* Always check element, not index
