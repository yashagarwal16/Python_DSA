# 🎯 Binary Search (Complete Interview Guide)

## 🧠 Problem Statement

Given a **sorted array `nums`** and a `target`, return the **index** of target.  
If not found → return `-1`.

---

## 🧪 Input

```python
nums = [2, 4, 6, 7, 9, 11, 18, 19]
target = 7
````

---

## 📦 Visual

```
Index:  0   1   2   3   4   5   6   7
Value:  2   4   6   7   9  11  18  19
```

---

## ✅ Output

```python
3
```

---

# 🧠 0. INTUITION (MOST IMPORTANT)

## 💡 Idea

👉 Binary Search works on **sorted data**

Instead of checking every element:

```
Linear Search → O(n)
Binary Search → O(log n)
```

👉 Core idea:

```
Check middle element → eliminate half of array
```

---

## 🔥 Real-Life Analogy

📖 Searching a word in dictionary:

* Open middle page
* If word is smaller → go left
* If bigger → go right

👉 You **cut search space in half every time**

---



# 🧱 1. ITERATIVE SOLUTION (Brute Force)

## 💡 Idea

Maintain:

```
low = 0
high = n - 1
```

At every step:

```
mid = (low + high) // 2
```

Compare:

* If `nums[mid] == target` → FOUND
* If smaller → go right
* If larger → go left

👉 Pattern: **Divide & Conquer (Iterative)**

--

`````python
def binarySearch(nums, target):
    low = 0
    high = len(nums) - 1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1
`````

`````python
nums = [2, 4, 6, 7, 9, 11, 18, 19]
target = 7
`````

`````python
binarySearch(nums, target)
`````

## 🔍 Dry Run

### Input:

```
nums = [2,4,6,7,9,11,18,19]
target = 7
```

---

### Step 1:

```
low=0, high=7
mid=(0+7)//2 = 3
nums[mid]=7 ✅ FOUND
```

👉 Return `3`

---

### Another Case (target = 13)

```
low=0, high=7
mid=3 → value=7 < 13 → go right

low=4, high=7
mid=5 → value=11 < 13 → go right

low=6, high=7
mid=6 → value=18 > 13 → go left

low=6, high=5 ❌ stop
```

👉 Return `-1`

---

## ⏱ Complexity

### Time: `O(log n)`

👉 WHY?

Each step:

```
n → n/2 → n/4 → n/8 → ... → 1
```

This happens:

```
log₂(n) times
```

Example:

```
32 → 16 → 8 → 4 → 2 → 1
Steps = 5 = log₂(32)
```

---

### Space: `O(1)`

👉 No extra memory used

---

## 🚨 When to Use

Use Binary Search when:

* Array is **sorted**
* You need **fast lookup**
* Interview says:

  * "Optimize from O(n)"
  * "Can you do better than linear?"

# 🔁 2. RECURSIVE SOLUTION

## 💡 Idea

Same logic, but:

👉 Solve smaller problem recursively

---

`````python
def binarySearch(nums, low, high, target):
    if low > high:
        return -1

    mid = (low + high) // 2

    if nums[mid] == target:
        return mid
    elif nums[mid] < target:
        return binarySearch(nums, mid + 1, high, target)
    else:
        return binarySearch(nums, low, mid - 1, target)
`````

`````python
nums = [2, 4, 6, 7, 9, 11, 18, 19]
target = 7

binarySearch(nums, 0, len(nums), target)
`````



## 🔍 Dry Run (target = 13)

```
Call 1: low=0, high=7 → mid=3 → go right
Call 2: low=4, high=7 → mid=5 → go right
Call 3: low=6, high=7 → mid=6 → go left
Call 4: low=6, high=5 → stop
```

---

## ⏱ Complexity

### Time: `O(log n)`

👉 Same reasoning as iterative

---

### Space: `O(log n)`

👉 WHY?

Each recursive call adds to stack:

```
Depth = log₂(n)
```

# 🧠 Pattern Summary

| Approach  | Pattern Used     |
| --------- | ---------------- |
| Iterative | Divide & Conquer |
| Recursive | Recursion + D&C  |

# 🔥 Interview Tips

### ⚠️ 1. Overflow Safe Mid

Instead of:

```python
mid = (low + high) // 2
```

Use:

```python
mid = low + (high - low) // 2
```

👉 Prevents overflow (important in Java/C++)

---

### ⚠️ 2. Variations You MUST Know

Binary Search is NOT just this problem:

* First occurrence
* Last occurrence
* Lower bound / Upper bound
* Search in rotated array
* Peak element
* Infinite sorted array
