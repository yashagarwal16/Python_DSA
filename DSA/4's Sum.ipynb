# 🎯 4Sum Problem

---



## 🧠 Problem Statement

Given an array `nums` and an integer `target`, return **all unique quadruplets** such that:

```

nums[i] + nums[j] + nums[k] + nums[l] = target
(i ≠ j ≠ k ≠ l)

````

---

## 🧪 Input

```python
nums = [1, 0, -1, 0, -2, 2]
target = 0
````

---

## 📦 Visual

```
Index:  0   1   2   3   4   5
Value:  1   0  -1   0  -2   2
```

---

## ✅ Output

```python
[[-2, -1, 1, 2],
 [-2,  0, 0, 2],
 [-1,  0, 0, 1]]
```

---



# 🧱 1. BRUTE FORCE

## 💡 Idea

* Try **all possible quadruplets**
* Check if sum == target
* Use set to remove duplicates

👉 Pattern: **4 Nested Loops + Hashing**

---

`````python

nums = [1,0,-1,0,-2,2]
target = 0

n = len(nums)
res = set()

for i in range(n):
    for j in range(i+1, n):
        for k in range(j+1, n):
            for l in range(k+1, n):
                if nums[i] + nums[j] + nums[k] + nums[l] == target:
                    temp = [nums[i], nums[j], nums[k], nums[l]]
                    temp.sort()
                    res.add(tuple(temp))

print([list(x) for x in res])
`````




## 🔍 Dry Run (Key Case)

Pick:

```
i=0 (1), j=1 (0), k=2 (-1), l=5 (2)

sum = 1 + 0 + (-1) + 2 = 2 ❌
```

Pick:

```
i=0 (1), j=2 (-1), k=3 (0), l=4 (-2)

sum = 1 -1 + 0 -2 = -2 ❌
```

Pick:

```
i=2 (-1), j=3 (0), k=4 (-2), l=5 (2)

sum = -1 + 0 -2 + 2 = -1 ❌
```

Pick:

```
i=2 (-1), j=3 (0), k=1 (0), l=0 (1)  (valid ordering adjusted)

→ [-1,0,0,1] ✅
```

---

## ⏱ Complexity

### Time: `O(n⁴)`

* 4 loops → checking all quadruplets

### Space: `O(k)`

* storing unique answers

---

## 🚨 When to Use

* Only for **very small n**
* Mainly for **understanding**

---



# ⚡ 2. BETTER (Hashing)

## 💡 Idea

Fix two indices `i` and `j`, then:

```
remaining = target - (nums[i] + nums[j])
```

Now solve **2Sum using a set**

👉 Pattern: **2 loops + Hashing**

---

`````python
nums = [1,0,-1,0,-2,2]
target = 0

n = len(nums)
res = set()

for i in range(n):
    for j in range(i+1, n):
        seen = set()
        for k in range(j+1, n):
            fourth = target - (nums[i] + nums[j] + nums[k])
            
            if fourth in seen:
                temp = [nums[i], nums[j], nums[k], fourth]
                temp.sort()
                res.add(tuple(temp))
            
            seen.add(nums[k])

print([list(x) for x in res])
`````

## 🔍 Dry Run

Fix:

```
i = -1, j = 0
```

Now:

```
k = 0 → need 1 → not found
k = 1 → need 0 → found ✅

→ [-1,0,0,1]
```

---

## ⏱ Complexity

### Time: `O(n³)`

* 3 loops + O(1) lookup

👉 WHY?
Fix 2 elements → remaining becomes 2Sum (linear scan)

### Space: `O(n)`

* set used

---

## 🚨 When to Use

* When interviewer says:
  👉 “Reduce one loop”

# 🚀 3. OPTIMAL (Sorting + Two Pointer)

## 💡 Idea

👉 Sort array first

```
[-2, -1, 0, 0, 1, 2]
```

Now:

* Fix `i`
* Fix `j`
* Use two pointers:

  * `k = j+1`
  * `l = n-1`

👉 Pattern: **Sorting + Two Pointer (Extended 3Sum)**

---



`````python
def fourSum(nums, target):
    nums.sort()
    n = len(nums)
    ans = []

    for i in range(n):
        if i > 0 and nums[i] == nums[i-1]:
            continue

        for j in range(i+1, n):
            if j > i+1 and nums[j] == nums[j-1]:
                continue

            k = j + 1
            l = n - 1

            while k < l:
                total = nums[i] + nums[j] + nums[k] + nums[l]

                if total < target:
                    k += 1
                elif total > target:
                    l -= 1
                else:
                    ans.append([nums[i], nums[j], nums[k], nums[l]])
                    k += 1
                    l -= 1

                    # skip duplicates
                    while k < l and nums[k] == nums[k-1]:
                        k += 1
                    while k < l and nums[l] == nums[l+1]:
                        l -= 1

    return ans
`````

`````python
nums=[-2, -1, 0, 0, 1, 2]
fourSum(nums, 0)
`````


## 🔍 Dry Run

Sorted:

```
[-2, -1, 0, 0, 1, 2]
```

### Step:

Fix:

```
i = -2, j = -1
```

```
k=2 (0), l=5 (2)

sum = -2 + -1 + 0 + 2 = -1 → too small → k++
```

```
k=3 (0), l=5 (2)

sum = -2 + -1 + 0 + 2 = -1 → k++
```

```
k=4 (1), l=5 (2)

sum = -2 + -1 + 1 + 2 = 0 ✅
→ [-2,-1,1,2]
```

---

## ⏱ Complexity

### Time: `O(n³)`

* Outer 2 loops → `n²`
* Inner two pointer → `O(n)`

👉 WHY not O(n⁴)?
Because instead of 2 loops inside, we use linear scan (k & l move once)

### Space: `O(1)` (excluding output)

---

## 🚨 When to Use

👉 VERY IMPORTANT PATTERN

Use when:

* k-Sum problems (3Sum, 4Sum, kSum)
* Need **unique combinations**
* Array-based + target sum
* Sorting is allowed

---

# 🧠 Pattern Summary

| Approach | Pattern Used            |
| -------- | ----------------------- |
| Brute    | 4 loops                 |
| Better   | Hashing (reduce 1 loop) |
| Optimal  | Sorting + Two Pointer   |

---

# 🔥 Interview Tip

If interviewer says:

* “Generalize 3Sum → ?” → Think **4Sum**
* “Can you extend this?” → k-Sum recursion
* “Avoid duplicates?” → Sorting + skip logic

---

# 💯 Final Intuition

```
4Sum = 2 loops + 2Sum
3Sum = 1 loop + 2Sum

Reduce dimension step-by-step
```

👉 Always try to convert:

```
k-Sum → (k-2 loops) + Two Pointer
```
