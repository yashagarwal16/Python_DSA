# 🔍 Missing Number (0 to n)

---

## 🧠 Idea

👉 Given array of size `n` containing numbers from `0 → n`  
👉 Exactly **one number is missing**

---

#  Brute Force

## 🧠 Idea

- Check every number from `0 → n`
- If not present in array → return it

---

````python
def missing_brute(nums):
    n = len(nums)

    for i in range(n + 1):
        if i not in nums:
            return i
````

````python
nums = [0, 1]
missing_brute(nums)
````

````python
nums = [9,6,4,2,3,5,7,0,1]
missing_brute(nums)
````

## 🔍 Dry Run

```python
nums = [3, 0, 1]
```

```text
i=0 → present  
i=1 → present  
i=2 → NOT present → answer = 2
```

---

## ⏱️ Complexity

**Time: O(n²)**

* `in` takes O(n) inside loop

**Space: O(1)**

---

## 🧩 Pattern

* Brute Force
* Linear Search

---

# ⚡ Better (Hashing)

## 🧠 Idea

* Store frequency of numbers `0 → n`
* Missing one will have freq = 0

---

````python
def missing_hash(nums):
    n = len(nums)
    frq = {}

    for i in range(n + 1):
        frq[i] = 0

    for num in nums:
        frq[num] = 1

    for k, v in frq.items():
        if v == 0:
            return k
````

````python
nums = [9,6,4,2,3,5,7,0,1]
missing_hash(nums)
````

## 🔍 Dry Run

```python
nums = [3, 0, 1]
```

```text
frq = {0:1, 1:1, 2:0, 3:1}

→ missing = 2
```

---

## ⏱️ Complexity

**Time: O(n)**

* Two passes

**Space: O(n)**

* Hash map

---

## 🧩 Pattern

* Hashing


# 🔥 Optimal (Math - Sum Formula)

## 🧠 Idea

* Sum of numbers `0 → n`:

  ```
  n*(n+1)//2
  ```
* Subtract array sum
* Remaining = missing number

---

````python
def missing_optimal(nums):
    n = len(nums)

    total = n * (n + 1) // 2
    arr_sum = sum(nums)

    return total - arr_sum
````

````python
nums = [9,6,4,2,3,5,7,0,1]
missing_optimal(nums)
````




## 🔍 Dry Run

```python
nums = [3, 0, 1]
```

```text
n = 3

total = 3*4/2 = 6
arr_sum = 3+0+1 = 4

missing = 6 - 4 = 2
```

---

## ⏱️ Complexity

**Time: O(n)**

* Sum of array

**Space: O(1)**

* No extra space

---

## 🧩 Pattern

* Math
* Prefix Sum Idea

---

## 🎯 When to Use

* Numbers are in range `0 → n`
* Exactly one missing
* Want **O(1) space**

---

## ⚠️ Notes

* Works only if:

  * No duplicates
  * Exactly one number missing
