# 🎯 Search in Rotated Sorted Array (LC 33)

---



## 🧠 Problem Statement

Given a **rotated sorted array `nums`** and a `target`:

👉 Return index of target  
👉 If not found → return `-1`

---

## 🧪 Input

```python
nums = [11,15,20,1,4,5,6,8,9,10]
target = 8
````

---

## 📦 Visual

```id="u3r4g2"
Original Sorted:
[1,4,5,6,8,9,10,11,15,20]

Rotated:
[11,15,20,1,4,5,6,8,9,10]
         ↑ pivot
```

---

## ✅ Output

```python
7
```

---

# 🧠 0. INTUITION (MOST IMPORTANT)

## 💡 Core Idea

👉 Array is NOT fully sorted
BUT…

```id="8c5j9y"
At least ONE HALF is always sorted
```

---

## 🔥 Key Thinking

At every step:

1. Find mid
2. Identify which half is sorted
3. Check if target lies in that half
4. Move accordingly

---



# 🧱 1. BRUTE FORCE

## 💡 Idea

👉 Simply scan entire array

👉 Pattern: **Linear Search**

---

`````python
def search(nums, target):
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1
`````

`````python
nums = [11,15,20,1,4,5,6,8,9,10]
target = 8
search(nums, target)
`````

## 🔍 Dry Run

```id="h1s9x2"
nums = [11,15,20,1,4,5,6,8,9,10]
target = 8

→ found at index 7
```

---

## ⏱ Complexity

### Time: `O(n)`

### Space: `O(1)`

---

## 🚨 When to Use

* Very small input
* Quick fallback

---



# 🚀 2. OPTIMAL (Binary Search)

## 💡 Idea

👉 Modified Binary Search

👉 Pattern: **Binary Search + Condition Logic**

---

`````python
def search(nums, target):
    low, high = 0, len(nums) - 1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == target:
            return mid

        # LEFT half sorted
        if nums[low] <= nums[mid]:
            if nums[low] <= target < nums[mid]:
                high = mid - 1
            else:
                low = mid + 1

        # RIGHT half sorted
        else:
            if nums[mid] < target <= nums[high]:
                low = mid + 1
            else:
                high = mid - 1

    return -1
`````

`````python
nums = [11,15,20,1,4,5,6,8,9,10]
target = 8
search(nums, target)
`````


## 🔍 Dry Run

### target = 8

```id="7u1v2g"
low=0, high=9
mid=4 → nums[4]=4

LEFT sorted? 11 ≤ 4 ❌ → RIGHT sorted

Check: 4 < 8 ≤ 10 ✅
→ low = mid+1 = 5
```

---

```id="h3w8k1"
low=5, high=9
mid=7 → nums[7]=8 ✅ FOUND
```

👉 Answer = 7

---

# ⏱ Complexity

### Time: `O(log n)`

👉 Same halving logic

### Space: `O(1)`
