# 🔍 Two Sum

---

# ❌ Your Code (Original Style)

## 🧠 Idea

- Fix `i`, move `j` ahead  
- Check every pair  

---

````python
def two_sum_my(nums, target):
    n = len(nums)

    for i in range(n-1):
        j = i + 1
        while j < n:
            total = nums[i] + nums[j]

            if total == target:
                return [i, j]

            j += 1
````

## 🔍 Dry Run

```python
nums = [5,9,1,2,4,15,6,3]
target = 13
```

```text
i=0 → check all j  
i=1 → 9 + 4 = 13 ✅ → return [1,4]
```

---

## ⏱️ Complexity

**Time: O(n²)**

* Nested traversal

**Space: O(1)**

---

## 🧩 Pattern

* Nested Loop

---

## 🎯 When to Use

* Basic brute logic
* Small inputs

# ✅ Standard Brute

## 🧠 Idea

* Use two loops
* Check all pairs

---

````python
def two_sum_brute(nums, target):
    n = len(nums)

    for i in range(n):
        for j in range(i+1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
````

## 🔍 Dry Run

```python
nums = [5,9,1,2,4,15,6,3]
target = 13
```

```text
i=1 → 9 + 4 = 13 → return [1,4]
```

---

## ⏱️ Complexity

**Time: O(n²)**

* All pairs checked

**Space: O(1)**

---

## 🧩 Pattern

* Nested Loop

---

## 🎯 When to Use

* Simple brute approach
* When optimization not required

# 🔥 Optimal (Hashing)

## 🧠 Idea

* Store visited numbers
* Check if complement exists

---

````python
def two_sum_optimal(nums, target):
    hashmap = {}

    for i in range(len(nums)):
        remaining = target - nums[i]

        if remaining in hashmap:
            return [hashmap[remaining], i]

        hashmap[nums[i]] = i
````

## 🔍 Dry Run

```python
nums = [5,9,1,2,4,15,6,3]
target = 13
```

```text
i=0 → 5 → store {5:0}  
i=1 → 9 → store  
i=2 → 1 → store  
i=3 → 2 → store  
i=4 → 4 → need 9 → found → return [1,4]
```

---

## ⏱️ Complexity

**Time: O(n)**

* Single pass

**Space: O(n)**

* Hashmap

---

## 🧩 Pattern

* Hashing

---

## 🎯 When to Use

* Interview optimal solution
* Unsorted array
* Need fast lookup
