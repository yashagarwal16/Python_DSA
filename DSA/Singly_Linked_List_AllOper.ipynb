## 📦 Structure of Node

````python
class Node:
    def __init__(self, val):
        self.val = val      # stores data
        self.next = None    # pointer to next node
````


## 🔥 Visualization

```
[5 | • ] → [10 | • ] → [7 | • ] → None
```

---

## ❓ WHY `next = None`?

* Marks end of list
* Prevents garbage traversal
* Default: node is not connected

# 🔁 2. TRAVERSAL

---

## 🧠 INTUITION

👉 Start from head → keep moving until None

---
## 🔥 Visualization

```
HEAD → 5 → 10 → 7 → None
         ↑
      current
```

---

````python
def traversal(self):
    if self.head is None:
        print("Linked List is empty")
        return

    current = self.head

    # WHY: visit every node including last
    while current is not None:
        print(current.val, end=" -> ")
        current = current.next

    print("None")
````

## 🔍 DRY RUN

List:

```
5 → 10 → 7 → None
```

Steps:

```
current = 5 → print 5
current = 10 → print 10
current = 7 → print 7
current = None → stop
```

---

## ⏱ COMPLEXITY

* Time: **O(n)** → visiting all nodes
* Space: **O(1)**

---

## 🚨 EDGE CASE

* Empty list → prints nothing

# ➕ 3. APPEND (INSERT AT END)

---

## 🧠 INTUITION

👉 Find last node → attach new node

---

## 🔥 Visualization

```
Before:
5 → 10 → 7 → None

After append(8):
5 → 10 → 7 → 8 → None
```

---

````python
def append(self, val):
    new_node = Node(val)

    # WHY: empty list
    if self.head is None:
        self.head = new_node
        return

    current = self.head

    # WHY: reach last node
    while current.next is not None:
        current = current.next

    # attach
    current.next = new_node
````

## 🔍 DRY RUN

Before:

```
5 → 10 → 7 → None
```

Append(8)

Steps:

```
current = 5 → 10 → 7
stop at last
7.next = 8
```

After:

```
5 → 10 → 7 → 8 → None
```

---
## ⏱ COMPLEXITY

* Time: **O(n)**
* Space: **O(1)**

---

## 🚨 EDGE CASE

* Empty list → becomes head

# 📍 4. INSERT AT SPECIFIC POSITION

---

## 🧠 INTUITION

👉 Reach position → insert between nodes
👉 Use **prev + current**

---

## 🧠 CORE IDEA

👉 Don’t move nodes
👉 Change links

---

## 🔥 Visualization

```
Insert 100 at pos = 2

Before:
5 → 10 → 7 → 8

After:
5 → 10 → 100 → 7 → 8
```

---

## ❓ WHY 2 POINTERS?

* prev → previous node
* current → next node

---

## ⚙️ HOW IT WORKS

```
prev.next = new_node
new_node.next = current
```

---

````python
def insert_at_specific(self, val, position):
    new_node = Node(val)

    # CASE 1: insert at head
    if position == 0:
        new_node.next = self.head
        self.head = new_node
        return

    current = self.head
    prev = None
    count = 0

    # traverse to position
    while current is not None and count < position:
        prev = current
        current = current.next
        count += 1

    # connect links
    prev.next = new_node
    new_node.next = current
````



## 🔍 DRY RUN

Insert 100 at position 2

Before:

```
5 → 10 → 7 → 8
```

Steps:

```
count=0 → current=5
count=1 → current=10
count=2 → stop

prev = 10
current = 7
```

Link:

```
10 → 100 → 7
```

After:

```
5 → 10 → 100 → 7 → 8
```

---

## ⏱ COMPLEXITY

* Time: **O(n)**
* Space: **O(1)**

---

## 🚨 EDGE CASES

* position = 0 → insert at head
* position > length → insert at end

---

## ❌ COMMON MISTAKE

```
Forget: new_node.next = current ❌
```

👉 breaks list

# 🗑 5. DELETE NODE

---

## 🧠 INTUITION

👉 Skip node
👉 Connect previous to next

---

## 🔥 Visualization

```
Delete 10

Before:
5 → 10 → 7 → 8

After:
5 → 7 → 8
```

---

## ⚙️ HOW

```
prev.next = current.next
```

---

````python
def delete(self, val):
    current = self.head

    # delete head
    if current and current.val == val:
        self.head = current.next
        return

    prev = None

    # find node
    while current and current.val != val:
        prev = current
        current = current.next

    if current is None:
        print("Value not found")
        return

    # skip node
    prev.next = current.next
````


## ⏱ COMPLEXITY

* Time: **O(n)**
* Space: **O(1)**

## 🚨 EDGE CASES

* delete head
* value not found
* single node

````python
# 🔥 COMPLETE FINAL STRUCTURE (ALL TOGETHER)


class Node:
    def __init__(self, val):
        self.val = val
        self.next = None


class SinglylinkedList:
    def __init__(self):
        self.head = None

    def traversal(self):
        current = self.head
        while current is not None:
            print(current.val, end=" -> ")
            current = current.next
        print("None")

    def append(self, val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            return

        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node

    def insert_at_specific(self, val, position):
        new_node = Node(val)

        if position == 0:
            new_node.next = self.head
            self.head = new_node
            return

        current = self.head
        prev = None
        count = 0

        while current is not None and count < position:
            prev = current
            current = current.next
            count += 1

        prev.next = new_node
        new_node.next = current

    def delete(self, val):
        current = self.head

        if current and current.val == val:
            self.head = current.next
            return

        prev = None
        while current and current.val != val:
            prev = current
            current = current.next

        if current:
            prev.next = current.next

````
