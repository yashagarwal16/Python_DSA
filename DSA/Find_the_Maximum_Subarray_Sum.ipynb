# 📈 Maximum Subarray Sum (Kadane’s Algorithm)

---

# Brute Force

## 🧠 Idea

- Start from each index `i`  
- Keep adding elements till end  
- Track maximum sum  

---

````python
def max_subarray_brute(nums):
    n = len(nums)
    maxi = float("-inf")

    for i in range(n):
        total = 0
        for j in range(i, n):
            total += nums[j]
            maxi = max(maxi, total)

    return maxi
````

````python
nums = [-2,1,-3,4,-1,2,1,-5,4]
max_subarray_brute(nums)
````

# 🔥 Optimal (Kadane’s Algorithm)

## 🧠 Idea

* Keep running sum
* If sum becomes negative → reset
* Track maximum

---

````python
def max_subarray_optimal(nums):
    maxi = float("-inf")
    total = 0

    for i in range(len(nums)):
        total += nums[i]
        maxi = max(maxi, total)

        if total < 0:
            total = 0

    return maxi
````

````python
nums = [-2,1,-3,4,-1,2,1,-5,4]
max_subarray_optimal(nums)
````

## ⏱️ Complexity

* Single pass

## 🧩 Pattern

## 🎯 When to Use

* Handles negative values
