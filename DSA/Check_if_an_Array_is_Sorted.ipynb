# 🔍 Check if an Array is Sorted

## 📌 Problem
Check whether a given array is **sorted in ascending order**

---

## 🧠 Idea

- Traverse the array  
- Compare each element with the next one  
- If any element is greater than next → **Not sorted**  

---



````python
def is_sorted(nums):
    for i in range(0, len(nums) - 1):
        if nums[i] > nums[i + 1]:
            return False
    return True
````

````python
nums = [1, 3, 2, 4, 5, 2, 4]
print(is_sorted(nums))
````

👉 Reason: 3 > 2 → Not sorted

````python
nums = [1, 1, 2, 3, 4, 7, 9]
print(is_sorted(nums))
````

👉 Array is in ascending order

## ⚡ Complexity

* Time Complexity → **O(n)**
* Space Complexity → **O(1)**

---

## 💡 Edge Cases

```python id="f7eqp3"
[] → True (empty array is sorted)
[5] → True (single element)
[5, 5, 5] → True (duplicates allowed)
```

---

## ✨ Summary

> Array is sorted if **every element ≤ next element**
