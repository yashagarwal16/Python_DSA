# 🔃 Sorting

## 📖 What is Sorting?

→ **Sorting** is the process of arranging elements in a specific order:
- Ascending (small → large)
- Descending (large → small)

---

## 🧠 Why Sorting?

- Makes searching faster 🔍  
- Organizes data efficiently  
- Used in many algorithms  

# 🟢 Selection Sort

## 📖 What is Selection Sort?

→ Selection Sort repeatedly:
1. Finds the **minimum element** from the unsorted part  
2. Swaps it with the current index  

---

## 🧠 Idea

- Assume index `i` is the minimum  
- Traverse from `i+1` to `n`
- Find actual minimum element  
- Swap with `i`  

---

## 🔄 Example

```python
nums = [1, 3, 2, 4, 5, 8, 9]

→ Steps:

- i = 0 → find min → swap
- i = 1 → find min → swap
continue till array is sorted

````python
def selection_sort(nums):
    n = len(nums)

    for i in range(n):
        min_index = i

        for j in range(i + 1, n):
            if nums[j] < nums[min_index]:
                min_index = j

        nums[i], nums[min_index] = nums[min_index], nums[i]
````

````python
# ⚡ Time Complexity (TC)
# Outer loop runs → n times
# Inner loop runs → (n-1), (n-2), ..., 1 times
# Total comparisons → n + (n-1) + (n-2) + ... + 1 = n(n-1)/2 → O(n^2)
# 👉 TC = O(n²)
          


# ⚡ Space Complexity (SC)
# In-place sorting → O(1) (no extra space used)
````

# 🔃 Bubble Sort

## 📖 What is Bubble Sort?

→ **Bubble Sort** is a sorting algorithm that:
- Compares **adjacent elements**
- Swaps them if they are in the wrong order  

👉 Largest elements "bubble up" to the end

---

## 🧠 Idea

- Compare adjacent elements:
  - If left > right → swap  
- Repeat for the whole array  
- After each pass, the largest element moves to its correct position  

---

## 🔄 Example

```python
nums = [5, 8, 1, 6, 9, 2, 4]

Pass 1:
(5,8) ✔️
(8,1) ❌ → swap → [5,1,8,...]
(8,6) ❌ → swap
(9 moves to end)

````python
def bubble_sort(nums):
    n = len(nums)

    for i in range(n - 1, 0, -1):   # controls passes
        for j in range(0, i):       # adjacent comparison
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
````

````python
# # # # Time Complexity (TC)
# # # Total comparisons:
# # # (n-1) + (n-2) + ... + 1 = n(n-1)/2

# # # 👉 TC = O(n²)

# # Cases:
# # Worst Case → O(n²)
# # Average Case → O(n²)

# 💾 Space Complexity (SC)
# No extra space used
# 👉 SC = O(1)
````

````python
'''
for best case O(n) when the array is already sorted
 and we can optimize the algorithm by adding a flag 
 to check if any swaps were made during a pass. 
 If no swaps were made, the array is already sorted, 
 and we can break out of the loop early.

'''

## SO 

def bubble_sort(nums):
    n = len(nums)

    for i in range(n - 1, 0, -1):   # controls passes
        is_swapped = False  # Flag to check if a swap occurred
        for j in range(0, i):       # adjacent comparison
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
                is_swapped = True
        if not is_swapped:
            break
# # # Time Complexity (TC)
# # # Best Case → O(n) (when the array is already sorted)
````
