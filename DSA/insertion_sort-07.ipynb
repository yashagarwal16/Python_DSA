# 🔃 Insertion Sort

## 📖 What is Insertion Sort?

→ **Insertion Sort** works like sorting playing cards in your hand 🃏  
→ It builds the sorted array **one element at a time**

---

## 🧠 Idea

- Assume first element is already sorted  
- Pick next element (`key`)  
- Compare it with previous elements  
- Shift larger elements to the right  
- Insert `key` at correct position  

---

## 🔄 Example

```python
nums = [3, 4, 3, 4, 5, 6, 3, 67, 2, 9]

Steps:

Take 4 → already sorted with 3 ✔️
Take 3 → shift 4 → insert 3
Take next → keep inserting in correct position

````python
def insertion_sort(nums):
    n = len(nums)

    for i in range(1, n):
        key = nums[i]
        j = i - 1

        while j >= 0 and nums[j] > key:
            nums[j + 1] = nums[j]
            j -= 1

        nums[j + 1] = key
````

````python
⚡ Time Complexity (TC)
Worst Case:
Reverse sorted array
All elements shifted

👉 O(n²)

Average Case:

👉 O(n²)

Best Case:
Already sorted

👉 O(n)

💾 Space Complexity (SC)
No extra space used
👉 O(1)
💡 Key Points
Efficient for small datasets
Works well for nearly sorted arrays
Stable sorting algorithm
Used in real-world systems (hybrid sorts)
✨ Summary

Insertion Sort = Pick element → Shift → Insert at correct position


---

## 🔥 Quick Interview Line

> “Insertion sort builds a sorted array one element at a time by inserting each element into its correct position.”

---

Now you have:
- ✅ Selection Sort  
- ✅ Bubble Sort  
- ✅ Insertion Sort  

If you want, I can give:
🔥 **Comparison Table (VERY IMPORTANT for exams/interviews)** 👍
````
