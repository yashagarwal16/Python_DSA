

# 🔀 Rearrange Array by Sign (Alternate + and -)





#  Brute Force

## 🧠 Idea

- Separate positives and negatives  
- Place them alternately in original array  

---

````python
def rearrange_brute(nums):
    pos = []
    neg = []

    for num in nums:
        if num >= 0:
            pos.append(num)
        else:
            neg.append(num)

    for i in range(len(pos)):
        nums[2*i] = pos[i]
        nums[2*i + 1] = neg[i]

    return nums
````

````python
nums = [5,10,-3,-1,-10,6]
rearrange_brute(nums)
````

## ⏱️ Complexity

**Time: O(n)**

* One pass to split + one to merge

**Space: O(n)**

* Extra arrays

---

## 🧩 Pattern

* Array splitting
* Index mapping

---

## 🎯 When to Use

* When extra space allowed
* Easy implementation

# 🔥 Optimal

## 🧠 Idea

* Create result array
* Place positives at even indices (0,2,4...)
* Place negatives at odd indices (1,3,5...)

---

````python
def rearrange_optimal(nums):
    n = len(nums)
    result = [0] * n

    posIndex = 0
    negIndex = 1

    for num in nums:
        if num >= 0:
            result[posIndex] = num
            posIndex += 2
        else:
            result[negIndex] = num
            negIndex += 2

    return result
````

````python
nums = [5,10,-3,-1,-10,6]
rearrange_optimal(nums)
````
