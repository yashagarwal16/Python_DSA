# 🎯 Search in Rotated Sorted Array II (LC 81)

## 🧠 Problem Twist

👉 SAME as LC-33

## 🚨 Why Normal (LC-33) Code FAILS?

### 🔥 Key Problem

```python
if nums[low] <= nums[mid]:
````

👉 This line assumes we can detect sorted half

❗ But with duplicates:

```python id="p6q8x1"
nums = [7,7,7,7,7,7,7,1,2,3,4,5,7,7]
         L       M               H
```

👉 Here:

```python id="wq9z2m"
nums[low] == nums[mid] == nums[high]
```

😵 We CANNOT decide:

* left sorted?
* right sorted?

👉 BOTH look same → **decision breaks**

---

# ❌ Why Previous Code Fails

## ⚠️ Ambiguous Case

```python id="9v5h2x"
nums[low] <= nums[mid]   ✔ TRUE
nums[mid] <= nums[high] ✔ TRUE
```

👉 Both halves look sorted
👉 But actually rotation exists inside

👉 So algorithm may:

* go wrong direction
* miss target

---

# 🧠 FIX (IMPORTANT)

## 💡 Idea

👉 When duplicates block decision:

```python id="5n1y8a"
nums[low] == nums[mid] == nums[high]
```

👉 Just shrink space:

```python id="y4c8pa"
low += 1
high -= 1
```

---


# 🚀 OPTIMAL CODE

`````python
def search(nums, target):
    low, high = 0, len(nums) - 1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == target:
            return True

        # ❗ duplicate case (main fix)
        if nums[low] == nums[mid] == nums[high]:
            low += 1
            high -= 1
            continue

        # LEFT sorted
        if nums[low] <= nums[mid]:
            if nums[low] <= target < nums[mid]:
                high = mid - 1
            else:
                low = mid + 1

        # RIGHT sorted
        else:
            if nums[mid] < target <= nums[high]:
                low = mid + 1
            else:
                high = mid - 1

    return False
`````

`````python
nums = [7,7,7,7,7,7,7,1,2,3,4,5,7,7]
search(nums, 1)
`````

`````python
search(nums, 8)
`````

# 🔍 Why This Fix Works

## 💡 Intuition

👉 When all three equal:

```python id="c3x91q"
nums[low] == nums[mid] == nums[high]
```

👉 We gain NO info

So:

```python id="7bq4ye"
just remove duplicates from edges
```
