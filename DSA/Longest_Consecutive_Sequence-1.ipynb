# 📏 Longest Consecutive Sequence


# Brute Force

## 🧠 Idea

- For each number, keep checking if `num+1` exists  
- Count length of sequence  

---


````python
def longest_brute(nums):
    max_count = 0
    n = len(nums)

    for i in range(n):
        count = 1
        num = nums[i]

        while num + 1 in nums:
            count += 1
            num += 1

        max_count = max(max_count, count)

    return max_count
````


---

## ⏱️ Complexity

**Time: O(n²)**

* `in` check is O(n) inside loop

**Space: O(1)**

---

## 🧩 Pattern

* Brute Force
* Linear Search

# ⚡ Better (Sorting)

## 🧠 Idea

* Sort array
* Count consecutive elements

---

````python
def longest_better(nums):
    nums = sorted(nums)
    n = len(nums)

    longest = 1
    count = 1

    for i in range(1, n):
        if nums[i] == nums[i-1]:
            continue
        elif nums[i] == nums[i-1] + 1:
            count += 1
        else:
            count = 1

        longest = max(longest, count)

    return longest
````

# 🔥 Optimal (Set)

## 🧠 Idea

* Use set for O(1) lookup
* Start only when `num-1` not present
* Expand sequence

---

````python
def longest_optimal(nums):
    myset = set(nums)
    longest = 0

    for num in myset:
        if num - 1 not in myset:
            count = 1
            x = num

            while x + 1 in myset:
                count += 1
                x += 1

            longest = max(longest, count)

    return longest
````

## ⏱️ Complexity

**Time: O(n)**

* Each number processed once

**Space: O(n)**

* Set

---

## 🧩 Pattern

* Hashing (Set)
* Sequence expansion

---

## 🎯 When to Use

* Consecutive sequence problems
* Need O(n) solution
* Unsorted array
