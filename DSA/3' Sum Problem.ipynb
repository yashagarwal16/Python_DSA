# 🎯 3' Sum Problem

---

## 🧠 Problem Statement

Given an array `arr`, find **all unique triplets** such that:

```

arr[i] + arr[j] + arr[k] = 0
(i ≠ j ≠ k)

````

---

## 🧪 Input

```python
arr = [-1, 0, 1, 2, -1, -4]
````

---

## 📦 Visual

```
Index:  0   1   2   3   4   5
Value: -1   0   1   2  -1  -4
```

---

## ✅ Output

```python
[[-1, -1, 2], [-1, 0, 1]]
```

---




# 🧱 1. BRUTE FORCE

## 💡 Idea

* Try **all possible triplets**
* Check if sum = 0
* Use **set** to avoid duplicates

👉 Pattern: **3 Nested Loops + Hashing**

---

`````python
arr = [-1,0,1,2,-1,-4]

my_set = set()
n = len(arr)

for i in range(0, n):
    for j in range(i+1, n):
        for k in range(j+1, n):
            if arr[i] + arr[j] + arr[k] == 0:
                temp = [arr[i], arr[j], arr[k]]
                temp.sort()   # ❗ important
                my_set.add(tuple(temp))

print([list(i) for i in my_set])
`````

## 🔍 Dry Run (Key Case)

Pick:

```
i=0 (-1), j=1 (0), k=2 (1)

sum = -1 + 0 + 1 = 0 ✅
→ add [-1,0,1]
```

Pick:

```
i=0 (-1), j=2 (1), k=4 (-1)

sum = -1 + 1 + (-1) = -1 ❌
```

Pick:

```
i=0 (-1), j=3 (2), k=4 (-1)

sum = -1 + 2 + (-1) = 0 ✅
→ add [-1,-1,2]
```

---

## ⏱ Complexity

### Time: `O(n³)`

* 3 loops → checking all triplets

### Space: `O(k)`

* storing unique triplets in set

---

## 🚨 When to Use

* Only when constraints are **very small**
* Mostly for **understanding**

---



# ⚡ 2. BETTER (Hashing)

## 💡 Idea

Fix `i`, then:

* Use a **set** to find the 3rd element

```
third = -(arr[i] + arr[j])
```

👉 Pattern: **Hashing + Two loops**

---

`````python
arr = [-1,0,1,2,-1,-4]
n = len(arr)
result = set()

for i in range(n):
    my_set = set()
    for j in range(i+1, n):
        third = -(arr[i] + arr[j])
        
        if third in my_set:
            temp = [arr[i], arr[j], third]
            temp.sort()
            result.add(tuple(temp))
        
        my_set.add(arr[j])

print([list(x) for x in result])
`````

## 🔍 Dry Run

Fix `i = -1`

```
j = 0 → need 1 → not found
j = 1 → need 0 → found ✅
→ [-1,0,1]
```

---

## ⏱ Complexity

### Time: `O(n²)`

* 2 loops + O(1) lookup

### Space: `O(n)`

* set storage

---

## 🚨 When to Use

* When interviewer says:
  👉 “Improve brute force”

---



# 🚀 3. OPTIMAL (Two Pointer)


## 💡 Idea

👉 Sort array first

```
[-4, -1, -1, 0, 1, 2]
```

Fix `i`, then use:

* `j = i+1`
* `k = n-1`

Move pointers based on sum

👉 Pattern: **Sorting + Two Pointer**

---

`````python
def threeSum(nums):
    nums.sort()
    n = len(nums)
    ans = []

    for i in range(n):
        if i > 0 and nums[i] == nums[i-1]:
            continue

        j = i + 1
        k = n - 1

        while j < k:
            total = nums[i] + nums[j] + nums[k]

            if total < 0:
                j += 1
            elif total > 0:
                k -= 1
            else:
                ans.append([nums[i], nums[j], nums[k]])
                j += 1
                k -= 1

                # skip duplicates
                while j < k and nums[j] == nums[j-1]:
                    j += 1
                while j < k and nums[k] == nums[k+1]:
                    k -= 1

    return ans
`````

`````python
nums=[-4, -1, -1, 0, 1, 2]
threeSum(nums)
`````

## 🔍 Dry Run

Sorted:

```
[-4, -1, -1, 0, 1, 2]
```

### Step:

Fix `i = -1`

```
j=2 (-1), k=5 (2)

sum = -1 + (-1) + 2 = 0 ✅
→ add [-1,-1,2]
```

Next:

```
j=3 (0), k=4 (1)

sum = -1 + 0 + 1 = 0 ✅
→ add [-1,0,1]
```

---

## ⏱ Complexity

### Time: `O(n²)`

* Outer loop + two pointer scan

👉 WHY not O(n³)?
Because inner work is linear (j & k move once)

### Space: `O(1)` (excluding output)

---

## 🚨 When to Use

👉 VERY IMPORTANT INTERVIEW PATTERN

Use when:

* Array + target sum
* Need unique pairs/triplets
* Sorted helps

# 🧠 Pattern Summary

| Approach | Pattern Used        |
| -------- | ------------------- |
| Brute    | 3 loops             |
| Better   | Hashing             |
| Optimal  | Sorting + 2 Pointer |

```
Brute → Try everything
```
