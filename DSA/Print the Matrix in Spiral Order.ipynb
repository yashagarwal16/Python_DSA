# 🌀 Spiral Matrix Traversal

---


## 🧠 Idea

👉 Traverse matrix in spiral order using **4 boundaries**:

- `top` → starting row  
- `bottom` → ending row  
- `left` → starting column  
- `right` → ending column  

👉 After each traversal → shrink boundaries

# 🧠 Why These Variables?

```python
top = 0        # first row
bottom = n-1   # last row
left = 0       # first column
right = m-1    # last column
```

👉 They define the **current boundary of traversal**

---

# 🔄 Traversal Steps

---

## 1️⃣ Left → Right (Top Row)

```python
for i in range(left, right+1):
    print(matrix[top][i])
top += 1
```

👉 Why?

* Fix row = `top`
* Move across columns

---

## 2️⃣ Top → Bottom (Right Column)

```python
for i in range(top, bottom+1):
    print(matrix[i][right])
right -= 1
```

👉 Why?

* Fix column = `right`
* Move down rows

---

## 3️⃣ Right → Left (Bottom Row)

```python
for i in range(right, left-1, -1):
    print(matrix[bottom][i])
bottom -= 1
```

👉 Why?

* Fix row = `bottom`
* Move backwards

---

## 4️⃣ Bottom → Top (Left Column)

```python
for i in range(bottom, top-1, -1):
    print(matrix[i][left])
left += 1
```

👉 Why?

* Fix column = `left`
* Move upwards

---



`````python

def spiralOrder(matrix):
    if not matrix or not matrix[0]:
        return []

    result = []

    top, left = 0, 0
    bottom, right = len(matrix)-1, len(matrix[0])-1

    while top <= bottom and left <= right:

        # 1. Left → Right
        for i in range(left, right+1):
            result.append(matrix[top][i])
        top += 1

        # 2. Top → Bottom
        for i in range(top, bottom+1):
            result.append(matrix[i][right])
        right -= 1

        # 3. Right → Left
        if top <= bottom:
            for i in range(right, left-1, -1):
                result.append(matrix[bottom][i])
            bottom -= 1

        # 4. Bottom → Top
        if left <= right:
            for i in range(bottom, top-1, -1):
                result.append(matrix[i][left])
            left += 1

    return result
`````

`````python
matrix = [
    [1,2,3,4,5,6],
    [20,21,22,23,24,7],
    [19,32,33,34,25,8],
    [18,31,36,35,26,9],
    [17,30,29,28,27,10],
    [16,15,14,13,12,11]
]
print(spiralOrder(matrix))
`````

# 🔍 Dry Run (First Layer)

# ⏱️ Complexity

**Time: O(n × m)**

* Visit each element once

**Space: O(1)** (excluding result)

---

# 🧩 Pattern

* Matrix traversal
* Boundary shrinking

---

# 🎯 When to Use

* Spiral / layer traversal
* Matrix boundary problems
* Printing patterns

---

## ⚠️ Notes

* Always check:

  ```
  if top <= bottom
  if left <= right
  ```
* Avoid duplicate traversal
