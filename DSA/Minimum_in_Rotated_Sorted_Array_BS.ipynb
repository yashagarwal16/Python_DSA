# 🎯 Find Minimum in Rotated Sorted Array

---



## 🧠 Problem Statement

Given a **rotated sorted array (no duplicates)**:

👉 Find the **minimum element**

---

## 🧪 Input

```python
nums = [4,5,6,7,0,1,2]
````

---

## ✅ Output

```python id="w3k8v1"
0
```

---


# 🧱 1. BRUTE FORCE

## 💡 Idea

👉 Just scan and track minimum

---

`````python
def findMin(nums):
    mini = float('inf')

    for num in nums:
        mini = min(mini, num)

    return mini
`````

`````python
nums = [4,5,6,7,0,1,2]
findMin(nums)
`````

## ⏱ Complexity

* Time: `O(n)`
* Space: `O(1)`

---

# 🚀 2. OPTIMAL (Binary Search)

## 💡 CORE IDEA

👉 At least **one half is always sorted**

👉 Use that to eliminate half

---

# 🔥 BEST VERSION (CLEAN)

`````python
def findMin(nums):
    low, high = 0, len(nums) - 1

    while low < high:
        mid = (low + high) // 2

        # Right part sorted → min in left
        if nums[mid] <= nums[high]:
            high = mid
        else:
            low = mid + 1

    return nums[low]
`````

`````python
nums = [4,5,6,7,0,1,2]
findMin(nums)
`````

# 🧠 WHY THIS WORKS

## 🔍 Case 1

```python id="c7v2m1"
nums[mid] <= nums[high]
```

👉 Right side is sorted
👉 Minimum is in **left (including mid)**

So:

```python id="u2k8xv"
high = mid
```

---

## 🔍 Case 2

```python id="r5n1b3"
nums[mid] > nums[high]
```

👉 Minimum is in **right side**

So:

```python id="m1p7qs"
low = mid + 1
```

---

# 🔍 Dry Run

```python id="y8x3w2"
nums = [4,5,6,7,0,1,2]

low=0, high=6
mid=3 → nums[3]=7 > nums[6]=2 → go right

low=4, high=6
mid=5 → nums[5]=1 <= nums[6]=2 → go left

low=4, high=5
mid=4 → nums[4]=0 <= nums[5]=1 → go left

low=4, high=4 → STOP
```

👉 Answer = nums[4] = 0

---

# ⏱ Complexity

* Time: `O(log n)`
* Space: `O(1)`

---

# 🧠 EDGE CASE

## Already Sorted

```python id="l7s2mn"
nums = [1,2,3,4]
```

👉 first element is min

Optimization:

```python id="q1m4zt"
if nums[low] <= nums[high]:
    return nums[low]
```

---

# 🔥 BONUS (WITH DUPLICATES)

👉 If duplicates allowed:

```python id="t8w3vn"
if nums[mid] == nums[high]:
    high -= 1
```

👉 Same ambiguity issue as LC-81

# 💯 FINAL INTUITION

```python id="x2m9ka"
Compare mid with high

→ shrink search space
```
