# ⚡ Quick Sort

## 📖 What is Quick Sort?

→ Quick Sort is a **Divide & Conquer** algorithm  

👉 Steps:
1. Pick a **pivot element**  
2. Place it in its **correct position**  
3. Elements smaller → left side  
4. Elements greater → right side  
5. Recursively apply same process  

---

## 🎯 What is Pivot?

→ A **pivot** is an element used to partition the array  

### 🔹 How to choose pivot?
- First element  
- Last element  
- Middle element  
- Random element (best in practice)

---

## 🧠 How it Works

Example:
```python
nums = [4, 6, 2, 5, 7, 1]
pivot = 4

After partition:
[2, 1]  4  [6, 5, 7]

👉 Pivot is now at correct position
👉 Left < pivot
👉 Right > pivot

🔁 Partition Logic
Start with two pointers:
i → from left
j → from right
Move:
i → until element > pivot
j → until element ≤ pivot
Swap when needed
Finally place pivot correctly

````python
def partition(nums, low, high):
    pivot = nums[low]
    i = low
    j = high

    while i < j:
        while i <= high - 1 and nums[i] <= pivot:
            i += 1
        
        while j >= low + 1 and nums[j] > pivot:
            j -= 1
        
        if i < j:
            nums[i], nums[j] = nums[j], nums[i]

    nums[low], nums[j] = nums[j], nums[low]
    return j
````

````python
# Quick Sort Function

def quick_sort(nums, low, high):
    if low < high:
        p_index = partition(nums, low, high)
        
        quick_sort(nums, low, p_index - 1)
        quick_sort(nums, p_index + 1, high)
````

⚡ Time Complexity (TC)
🟢 Best Case:
Pivot divides array into equal halves
T(n) = 2T(n/2) + O(n)

👉 O(n log n)

🟡 Average Case:
Balanced partitions (not perfect but close)

👉 O(n log n)

🔴 Worst Case:
Pivot is smallest or largest every time
Example: already sorted array
T(n) = T(n-1) + O(n)

👉 O(n²) ❌

❓ Why log(n)?
Each partition reduces problem size
Levels of recursion ≈ log(n) (best/avg)
Work per level = O(n)

👉 Total = O(n log n)

💾 Space Complexity (SC)
Due to recursion stack

👉 Best/Average → O(log n)
👉 Worst → O(n)

💡 Key Points
Faster than Merge Sort in practice
In-place sorting (no extra array)
Not stable
Performance depends on pivot choice
✨ Summary

Quick Sort = Pick Pivot → Partition → Recursively Sort

🔥 Interview One-Liner

“Quick sort partitions the array around a pivot and recursively sorts subarrays, achieving O(n log n) on average but O(n²) in worst case.”

## 🔥 Extra Insight (Very Important)

👉 Why Quick Sort is preferred over Merge Sort?

- Better **cache performance**
- No extra memory (in-place)
- Faster in real-world
