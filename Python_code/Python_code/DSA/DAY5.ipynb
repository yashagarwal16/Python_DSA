# 🔢 Factorial Using Recursion

## 📖 Definition
→ Factorial of a number **n** is:
- n! = n × (n-1) × (n-2) × ... × 1  
- Example: 5! = 120  

---

## 🧠 Recursive Idea
→ n! = n × (n-1)!  

---

## 🔑 Base Case
- 0! = 1  
- 1! = 1  

---

````python
def fact(n):
    if n == 0 or n == 1:
        return 1
    
    return n * fact(n - 1)
````

````python
# fact(5)
# = 5 × fact(4) \n
# = 5 × 4 × fact(3)
# = 5 × 4 × 3 × fact(2)
# = 5 × 4 × 3 × 2 × fact(1)
# = 120 -->
````

````python
print(fact(5))
````

# 🔁 Reverse an Array Using Recursion

## 📌 Problem
Reverse an array (or part of it) using recursion.

### Example:
```python
num = [5, 4, 3, 2, 4, 5, 4]
left = 2
right = 5

````python
def rev(arr, left, right):
    if left >= right:
        return
    
    # Swap
    arr[left], arr[right] = arr[right], arr[left]
    
    # Recursive call
    rev(arr, left + 1, right - 1)
````

````python
num = [5, 4, 3, 2, 4, 5, 4]

rev(num, 0, len(num) - 1)

print(num)
````

# 🔁 Reverse an Array Using While Loop

## 📌 Idea
- Use two pointers:
  - `left` → start index  
  - `right` → end index  
- Swap elements and move inward

---

````python
arr = [5, 4, 3, 2, 4, 5, 4]

left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print(arr)
````

# 🔁 Check if a String is Palindrome

## 📖 What is a Palindrome?

→ A **palindrome** is a string that reads the **same forward and backward**

### Examples:
- "madam" ✅  
- "racecar" ✅  
- "hello" ❌  

````python
### Approach 1: Using While Loop

def is_palindrome(s):
    left = 0
    right = len(s) - 1

    while left < right:
        if s[left] != s[right]:
            return False
        
        left += 1
        right -= 1
    
    return True
````

````python
is_palindrome("madam")
````

````python
is_palindrome("hello")
````

````python
### Approach 2: Using Recursion

def is_palindrome_rec(s, left, right):
    if left >= right:
        return True
    
    if s[left] != s[right]:
        return False
    
    return is_palindrome_rec(s, left + 1, right - 1)
````

````python
############### Complexity
# Time Complexity → O(n)
# Space Complexity:
# While loop → O(1)
# Recursion → O(n) (call stack)
````

# 🔢 Fibonacci Using Recursion

## 📖 What is Fibonacci?

→ A **Fibonacci sequence** is a series where:
- Each number is the **sum of previous two numbers**

### Example:
0, 1, 1, 2, 3, 5, 8, 13...

---

## 🧠 Recursive Flow

→ f(n) = f(n-1) + f(n-2)
f(7) = f(6) + f(5)
(f(5) + f(4)) + (f(4) + f(3))


---

## 🔑 Base Case

- f(0) = 0  
- f(1) = 1  

---

````python
def fib(n):
    if n == 0 or n == 1:
        return n
    
    return fib(n - 1) + fib(n - 2)
````

````python
print(fib(7))
````
