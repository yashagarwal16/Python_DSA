# 🔃 Merge Sort (Deep Understanding)

## 📖 What is Merge Sort?

→ Merge Sort is a **Divide & Conquer** algorithm:
1. Divide array into halves  
2. Sort each half  
3. Merge sorted halves  

---

## 🧠 How it Works (Step-by-Step)

Example:
```python
nums = [3, 1, 2, 4]

Step 1: Divide
[3,1,2,4]
→ [3,1]   [2,4]
→ [3] [1] [2] [4]

Step 2: Merge (Sort while merging)
[3] + [1] → [1,3]
[2] + [4] → [2,4]

[1,3] + [2,4] → [1,2,3,4]

🔥 Key Insight

👉 Splitting continues until single elements
👉 Merging builds the sorted array

````python
## 🔗 Merge Two Sorted Arrays

def merge(left, right):
    result = []
    i, j = 0, 0

    n, m = len(left), len(right)

    while i < n and j < m:
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Remaining elements
    while i < n:
        result.append(left[i])
        i += 1

    while j < m:
        result.append(right[j])
        j += 1

    return result
````

````python
## 🧩 Merge Sort Function

def merge_sort(arr):
    if len(arr) <= 1:
        return arr   # already sorted

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)
````

````python
nums = [3, 1, 2, 4, 1, 5, 2, 6, 4]

sorted_nums = merge_sort(nums)

print(sorted_nums)
````

⚡ Time Complexity (TC)
Divide → log n levels
Merge → O(n) work at each level

👉 TC = O(n log n)

Cases:
Best → O(n log n)
Average → O(n log n)
Worst → O(n log n)
💾 Space Complexity (SC)
Extra arrays used for merging

❓ Why Time Complexity = O(n log n)?
Step 1: How many times we divide?

Each time we divide array into half:

n → n/2 → n/4 → n/8 → ... → 1

👉 Number of divisions = log₂(n)

Step 2: Work done at each level

At each level, we merge all elements

👉 Total work per level = O(n)

👉 SC = O(n)

💡 Key Points
Based on Divide & Conquer
Stable sorting algorithm
Works well for large datasets
Requires extra memory
